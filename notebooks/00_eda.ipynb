### split multi answers

```python
def split_answer(answer , sep=';'):
    '''
    split answer by delimiter
    
    param:
    answer: pd.Series or pd.DataFrame 
    sep: delimiter(str) default ';'
    
    return:
    pd.Series or pd.DataFrame
    '''
    
    
    def check_delimiter(x):
        ''' check if delimiter exist'''
        return x.str.contains(sep)
    
    if check_delimiter(answer).any():
        modf =answer.str.split(sep)
        mask = modf.isna()
        modf[mask]=modf[mask].apply(lambda x: [])
        return modf
    else:
        return answer
    
    
    
```

```python
object_cols = df.select_dtypes(include='object').columns.tolist()
for col in object_cols:
    df[col] = split_answer(df[col])
```
